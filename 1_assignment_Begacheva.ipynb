# Installing packages and opening a dataset 

Installing necessary packages and loading a suitable dataset contained in the AER package - More Guns, Less Crime? Guns dataset is balanced panel of data on 50 US states, plus the District of Columbia (for a total of 51 states), by year for 1977–1999. It includes 1173 observations on 13 variables.

```r
library(data.table) 
install.packages('AER')
library(AER)
install.packages("DT") 
library(DT)
install.packages('tidyverse')
library(tidyverse)
```

```r
data("Guns")
setDT(Guns)
options(repr.matrix.max.rows=20, repr.matrix.max.cols=30)
```

```r
?Guns
```

```r
head(Guns)
```

# 1. Selecting colums 

```r
Guns %>% select(year, state, population, income)
```

# 2.Filtering out rows with violent crime rate higher than average

```r
mean(Guns$violent)
```

```r
Guns %>%
filter(violent > mean(Guns$violent))
```

# 3. Creating a new column "crime" using mutate operation as a sum of violent, murder, and robbery rates

```r
Guns %>%
mutate(crime = violent+ murder + robbery) 
```

# 4. Grouping the data by the state and summarizing its max crime rate

```r
Guns %>%
group_by(state) %>%
summarise(max_crime = max(violent+ murder + robbery))
```

# 5. Performing pivot operations

```r
Guns_at_wide <- Guns %>% pivot_wider(id_cols = state, names_from = year, values_from = violent)
Guns_at_wide
```

```r
Guns_at_wide %>% pivot_longer(cols= "state", names_to = "year", values_to = "violent")
```
